# Introduction to Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[10,8,6,9]
plt.plot(x,y,color='red')
```

### Pyplot API vs Object-Oriented(OO) API
* Pyplot API->quicly
* OO API->advanced

Pyplot API

```python
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x),color="blue",linestyle='dashed',label="sin(x)")
plt.plot(x,np.cos(x),color="red",label="cos(x)")
plt.xlabel("Bien X")
plt.ylabel("Sin(X)")
plt.title("Sin(x) and Cos(x)")
#plt.xlim([0,4])

#plt.axis([0,4,-0.7,0.6])
plt.axis("equal")
plt.legend();
```

### Object-Oriented API

```python
#1. Prepare data
x=[1,2,3,4]
y=[11,22,33,44]

#2. Setup plot
fig,ax=plt.subplots(figsize=(5,5))

#3. Plot the data
ax.plot(x,y)

#4. Customize
ax.set(title="A simple plot",
       xlabel="x-axis",
       ylabel="y-axis")
```

## 2.Most common types of Matplotlib plots
* line
* scatter
* bar
* hist
* subplots()

### Line

```python
# Create an array
x=np.linspace(0,10,100)
x[:5]
```

```python
#OO API plot a line chart

fig,ax=plt.subplots()
ax.plot(x,x**3)
```

## Scatter

```python
# Pyplot API

plt.scatter(x,np.exp(x))
```

```python
# OO API
#Prepare New Data
rng=np.random.RandomState(0)

x=rng.randn(100)
y=rng.randn(100)

colors=rng.rand(100)
sizes=1000*rng.rand(100)

fig,ax=plt.subplots()
img1=ax.scatter(x,y,s=sizes,c=colors,cmap='viridis',alpha=0.3)
fig.colorbar(img1)
```

### 3.Bar
* Vertical
* Horizontal

```python
#Prepare data from a python dictionary

soft_drink_prices={"Coke":10,"Pesi":12,"Sprite":8}

fig,ax=plt.subplots()
ax.bar(soft_drink_prices.keys(),soft_drink_prices.values())
ax.set(title="Bach Hoa Xanh's Soft Drink Prices",ylabel="Price ($)")
```

```python
fig,ax=plt.subplots()
ax.barh(list(soft_drink_prices.keys()),list(soft_drink_prices.values()))
```

## 4.Histogram

```python
#Prepare data

np.random.seed(42)
student_height=np.random.normal(170,10,250)
student_height[:10]
```

```python
#pyplot API
plt.hist(student_height,bins=30)
```

### Subplots


```python
#Option 1:
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))

#Plot data to each axis
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(soft_drink_prices.keys(),soft_drink_prices.values())
ax4.hist(student_height,bins=30)
```

```python
#Option 2:
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(soft_drink_prices.keys(),soft_drink_prices.values())
ax[1,1].hist(student_height,bins=30)
```
